```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = pd.read_csv("data/city_data.csv")
ride_data_to_load = pd.read_csv("data/ride_data.csv")

# Read the City and Ride Data

# Combine the data into a single dataset

# Display the data table for preview
```

```python
# Read the City and Ride Data
city_data_to_load.head()
```

```python
# Read the City and Ride Data
ride_data_to_load.head()
```

```python
# Combine the data into a single dataset
city_data = pd.merge(city_data_to_load, ride_data_to_load, on="city")
city_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
# Obtain the x and y coordinates for each of the three city types
urban_city_types = city_data[city_data["type"]== "Urban"].groupby(city_data["city"])
x_urban = urban_city_types["ride_id"].count()
y_urban = urban_city_types["fare"].mean()
z_urban = urban_city_types["driver_count"].mean()

urban = plt.scatter(x_urban,y_urban,marker="o",facecolors="red",
            edgecolors="black",s=z_urban,alpha=0.75)
urban
```

```python
rural_city_types = city_data[city_data["type"] == "Rural"].groupby("city")
x_rural = rural_city_types["ride_id"].count()
y_rural = rural_city_types["fare"].mean()
z_rural = rural_city_types["driver_count"].mean()

rural = plt.scatter(x_rural, y_rural,marker="o",facecolors="blue",
                   edgecolors="black",s=z_rural, alpha=0.75)
rural
```

```python
suburban_city_types = city_data[city_data["type"]=="Suburban"].groupby("city")
x_suburban = suburban_city_types["ride_id"].count()
y_suburban = suburban_city_types["fare"].mean()
z_suburban = suburban_city_types["driver_count"].mean()

suburban = plt.scatter(x_suburban,y_suburban,marker="o",facecolors="orange",
                      edgecolors="black",s=z_suburban,alpha="0.75")
suburban
```

```python
plt.scatter(x_urban,y_urban,marker="o",facecolors="red",
            edgecolors="black",s=z_urban*10,alpha=0.75, label= "Urban")
plt.scatter(x_rural, y_rural,marker="o",facecolors="blue",
                   edgecolors="black",s=z_rural*10, alpha=0.75, label= "Rural")
plt.scatter(x_suburban,y_suburban,marker="o",facecolors="orange", label= "Suburban",
                      edgecolors="black",s=z_suburban*10,alpha="0.75")

plt.legend()
plt.grid()
plt.xlabel("Total number of Rides per city")
plt.ylabel("Average Fare Cost")
plt.title("Pyber Ride Sharing")
plt.text(42,35,"Note: \nCircle size correlates with driver count per city.",fontsize = 12)

#plt.savefig("Images/PyberRideSharingData.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

# Build Pie Chart

# Save Figure
```

```python
types = city_data.groupby(['type'])
sum_fares = types['fare'].sum()
```

```python
label = ["Rural","Suburban","Urban"]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0,0,0.1)

plt.title("% of Total Fares by City Type")
plt.pie(sum_fares, explode=explode, labels=label, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
#plt.savefig("Images/TotalFaresbyCityType.png")
plt.axis("equal")
plt.show()
```

```python
# Show Figure
plt.show()
```
